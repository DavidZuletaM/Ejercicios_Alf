##Aplicar funciones sobre Dataframes

```python
#apply: Ejecutar funciones sobre un df, puede ser anónima o declarada
def elevar_al_cuadrado(edad_fraccion:float):
  return edad_fraccion**2
```

```python
df["edad_elevada"] = df["edad"].apply(elevar_al_cuadrado)
df
```

```python
#Funciones declaradas
df["edad_elevada_fraccion"] = df["edad_fraccion"].apply(elevar_al_cuadrado)
df
```

```python
#Funciones anónimas
df["edad_elevada_fraccion_2"] = df["edad_fraccion"].apply(lambda valor: valor**2)
df
```

```python
#Funciones usando varias columnas del df
def ajustar_ponderaciones (fila):
  fila["edad_ajustada"] = fila.constante + fila.edad_fraccion
  fila["nombre"] = fila["nombre"].upper()
  return fila
```

```python
#Para definir si la función se aplica a las columnas o a las filas:
#axis = 0 --- Aplicar por cada columna
#axis = 1 --- Aplicar por cada fila ---> Lo normal es que las funciones se apliquen a las filas

df.apply(ajustar_ponderaciones, axis=1)
```
